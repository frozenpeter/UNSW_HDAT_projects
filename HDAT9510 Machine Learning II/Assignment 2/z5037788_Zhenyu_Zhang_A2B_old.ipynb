### Import Libraries and Load Model Llama 3.2-1B:

[Use with transformers](https://huggingface.co/meta-llama/Llama-3.2-1B)

```python
import torch
from transformers import pipeline

# Specify the model ID
model_id = "meta-llama/Llama-3.2-1B"

# Load the pipeline for text generation
pipe = pipeline(
    "text-generation",
    model=model_id,
    max_new_tokens=50,
    do_sample=False,
    torch_dtype=torch.bfloat16,
    device_map="auto"
)
```

the Meta LLaMA-3.2-1B model has been downloaded and loaded successfully.

Files like config.json, model.safetensors, tokenizer_config.json, and others have been downloaded successfully. These files are essential components of the model, including the weights, tokenizer, and configuration.

The pipeline has been created (pipe) and can now be used for text generation

The warning about truncation (Truncation was not explicitly activated...) is informational and not an error. It means that the model is applying the default truncation strategy to handle long inputs.



Code explain:
- **pipeline**: A utility from Hugging Face's transformers library, which simplifies the process of loading pre-trained models and tokenizers and setting up a task-specific pipeline (e.g., text generation).
- `text-generation`: Specifies the task for the pipeline, which is generating text based on a given prompt.
model=model_id: Loads the pre-trained meta-llama/Llama-3.2-1B model for text generation.
- `max_new_tokens=50`: Limits the output to a maximum of 50 tokens to avoid excessively long or unwieldy results.
- `do_sample=False`: Disables sampling and makes the generation deterministic, ensuring the model selects the most probable token at each step.
  - Avoid using sampling-specific parameters (`temperature`, `top_p`, `top_k`) as they are ignored when do_sample=False.
- `torch_dtype=torch.bfloat16`: Optimizes memory usage by using the bfloat16 data type for computations, suitable for large models.
- `device_map="auto"`: Automatically detects and allocates the model to available hardware (e.g., GPU or CPU), optimizing execution.

## Solution to question 2 here:

```python
# Load the pipeline for text generation
pipe2 = pipeline(
    "text-generation",
    model=model_id,
    max_new_tokens=100,
    do_sample=False,
    torch_dtype=torch.bfloat16,
    device_map="auto"
)
```

```python
# Define health-related prompts
health_related_prompt_1 = "List the top five risk factors for cardiovascular diseases:"
health_related_prompt_2 = "Explain how lifestyle changes can help prevent cardiovascular diseases, focusing on diet and exercise:"
health_related_prompt_3 = "Describe the connection between smoking and cardiovascular diseases in detail, including mechanisms and statistics:"
```

```python
# Generate responses for each prompt
response_1 = pipe2(health_related_prompt_1)
response_2 = pipe2(health_related_prompt_2)
response_3 = pipe2(health_related_prompt_3)
```

```python
# Print the responses
print("Health-Related Prompt 1 Response:\n", response_1[0]['generated_text'], "\n")
print("Health-Related Prompt 2 Response:\n", response_2[0]['generated_text'], "\n")
print("Health-Related Prompt 3 Response:\n", response_3[0]['generated_text'], "\n")
```

### Compare the outputs and analyze how differences in prompt wording impact the response
---

#### **Health-Related Prompt 1**:
"List the top five risk factors for cardiovascular diseases:"

- **Output Analysis**:
  - The response misinterpreted the request, listing general medical terms (aetiology, prevention, diagnosis, treatment, and prognosis) instead of specific cardiovascular disease risk factors.
  - This indicates that the model struggled to understand the prompt's intent and reverted to a generic structure often associated with medical topics.

- **Impact of Prompt Wording**:
  - The phrasing of the prompt is too broad. The use of "risk factors" could have been clarified with examples (e.g., "List specific biological and behavioral risk factors"). The model's response suggests it may need more guidance to focus on specific types of information.

---

#### **Health-Related Prompt 2**:
"Explain how lifestyle changes can help prevent cardiovascular diseases, focusing on diet and exercise:"

- **Output Analysis**:
  - The response appears to offer a list of questions rather than providing a direct explanation of how lifestyle changes impact cardiovascular health.
  - The questions demonstrate that the model understands the prompt's theme but fails to address it directly. Instead, it outlines subtopics for discussion.

- **Impact of Prompt Wording**:
  - The phrase "explain how lifestyle changes can help" might have been too open-ended, leading the model to default to generating a structured outline instead of an explanatory paragraph.
  - Rewording to something like "Describe the benefits of diet and exercise in preventing cardiovascular diseases" might encourage a more focused and informative response.

---

#### **Health-Related Prompt 3**:
"Describe the connection between smoking and cardiovascular diseases in detail, including mechanisms and statistics:"

- **Output Analysis**:
  - The response is clear, detailed, and aligned with the prompt. It outlines the impact of smoking on cardiovascular diseases, mentioning specific mechanisms (e.g., increased risk of coronary heart disease and stroke) and includes statistics ("1 in 5 deaths in the United States is due to CVDs").
  - This output demonstrates the model's ability to generate informative and relevant content when the prompt is specific.

- **Impact of Prompt Wording**:
  - The specificity of this prompt significantly improved the response quality. The inclusion of keywords like "mechanisms" and "statistics" guided the model to address both qualitative and quantitative aspects, resulting in a more comprehensive answer.
